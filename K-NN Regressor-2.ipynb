```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from random import gauss
from sklearn.metrics import accuracy_score
from array import array
# generating the dataset

mean = [1, 0]
sigma = [[1, 0.75], [0.75, 1]]
X_dataframe = X, Y = np.random.multivariate_normal(mean, sigma, 300).T
test_dataframe = np.random.multivariate_normal(mean, sigma, 100).T
noise = gauss(0, 0.5)
df1 = pd.DataFrame(X_dataframe, index=['x', 'y']).T
df2 = pd.DataFrame(test_dataframe, index=['x', 'y']).T

df1["target"] = 2 * df1['x'] + df1['y'] + noise
df2["target"] = 2 * df2['x'] + df2['y'] + noise
test_dataframe1 = df2.values.tolist()
print(test_dataframe1)
# class knn
class KNN:

    def __init__(self, n):

        self.neighbors = n
        self.features = []
        self.m_target = []

    def knn_fit(self, x, y):

        self.features = x.values.tolist()
        self.m_target = y.values.tolist()

    def E_D(self, row_1, row_2):

        d = math.sqrt(sum([(a - b) ** 2 for a, b in zip(row_1, row_2)]))
        return d

    def voting_method(self, knn, vote_func):

        if vote_func == 'mean':
            sum_of_k_targets = 0
            for k, v in knn:
                sum_of_k_targets += v['target']
            return sum_of_k_targets / len(knn)

    def myknnregress(self, train, x, k, vote_func='mean'):

        distances = {}
        for i in range(len(self.features)):
            d = self.E_D(self.features[i], x)
            distances[i] = {'dist': d, 'target': self.m_target[i]}
        k_nearest_neighbors = sorted(distances.items(), key=lambda x: x[1]['dist'])[:self.neighbors]
        target_pred = self.voting_method(k_nearest_neighbors, vote_func)
        
        return target_pred


X_dataframe1 = df1.drop(["target"], axis=1)
Y_dataframe1 = df1["target"]

neighbors = [1, 2, 3, 4, 5, 10, 20, 50, 100]
for k in neighbors:
    knn_clf = KNN(k)
    knn_clf.knn_fit(X_dataframe1, Y_dataframe1)
    print("Accuracy for k:", k)
    x = 0
    for i in test_dataframe1:
        test = [i[0], i[1]]
        X_dataframe2 = knn_clf.myknnregress(X, test, k)
        actual_value = i[2]
        accuracy = 100 - (abs((actual_value - X_dataframe2) / actual_value) * 100)
        x = x + accuracy
    print(x / len(test_dataframe1)) 
    
```
