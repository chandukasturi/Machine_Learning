```python
import matplotlib.pyplot as plt
from math import sqrt
import numpy as np
import sys
import pickle
import pandas as pd
import random
from random import randrange
import statistics
from sklearn.model_selection import train_test_split



# generating the data set
mean1 = [1, 0]
mean2 = [0, 1]s
cov1 = [[1, 0.75], [0.75, 1]]
cov2 = [[1, -0.5], [0.5, 1]]
data_1 = x1, y1 = np.random.multivariate_normal(mean1, cov1, 200).T
data_2 = x2, y2 = np.random.multivariate_normal(mean2, cov2, 200).T
X = x, y = np.concatenate((data_1, data_2), axis=1)
df = pd.DataFrame(X, index=['x', 'y']).T
df['Class_Label'] = [random.randint(0, 1) for k in df.index]
X_dataframe = df.values.tolist()


data_3 = x3, y3 = np.random.multivariate_normal(mean1, cov1, 50).T
data_4 = x4, y4 = np.random.multivariate_normal(mean2, cov2, 50).T
X1 = x, y = np.concatenate((data_3, data_4), axis=1)
df1 = pd.DataFrame(X1, index=['x', 'y']).T
df1['Class_Label'] = [random.randint(0, 1) for k in df1.index]
test_dataframe = df1.values.tolist()



# Calculating  the Euclidean distance 
def E_D(r1, r2):
    d = 0.0
    for i in range(len(r1) - 1):
        d += (r1[i] - r2[i]) ** 2
    return sqrt(d)


# Finding the k nearest neighbors
def neighbours_k(X, test_dataframe, number_of_neighbours):
    distances = list()
    for X_row in X:
        dist = E_D(test_dataframe, X_row)
        distances.append((X_row, dist))
    distances.sort(key=lambda tup: tup[1])
    neighbors = list()
    for i in range(number_of_neighbours):
        neighbors.append(distances[i][0])
    return neighbors


# Predicting the class labels
def myknnclassify(X, test_dataframe, number_of_neighbours):
    #     for x in number_of_neighbours:
    neighbors = neighbours_k(X, test_dataframe, number_of_neighbours)
    output_values = [row[-1] for row in neighbors]
    prediction = max(set(output_values), key=output_values.count)
    
    return prediction


# Calculating accuracy 
def calculate_accuracy(actual, predicted, i):
    c = 0
    for x in range(len(i)):
        if actual == predicted:
            c += 1
    return c / float(len(i)) * 100.0



number_of_neighbours = [1, 2, 3, 4, 5, 10, 20]
# printing the accuracy
my_list = []
for j in number_of_neighbours:
    x = 0
    for i in test_dataframe:
        label = myknnclassify(X_dataframe, i, j)
        x = x + calculate_accuracy(i[2], label, i)

    print("Accuracy for k = ", j, x / len(test_dataframe))



```
